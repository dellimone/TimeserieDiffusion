## Sinusoidal positional embedding

$$
PE_{(pos,2i)} = \sin(pos/10000^{2i/d_{model}})
$$
$$
PE_{(pos,2i+1)} = \cos(pos/10000^{2i/d_{model}})
$$

Properties of Sinusoidal Positional Encoding:

- **Normalized Range**: The sine and cosine functions used in the encoding keep the values within the range of [-1,1], which is a normalized range that is easy for the model to learn.
- **Unique Position Encoding**: Each position is encoded uniquely, enabling the model to distinguish between different positions effectively. It also virtually scales to unlimited sequence lengths.
- **Relative distance**: The sinusoidal positional encoding maintains relative distance between positions. For example, the angle between the encodings of position 5 and 6 is equal to the angle of 6 and 7, which represents the relative distance between the positions.
